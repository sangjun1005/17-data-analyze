```python
# 행 인덱스 변경: DataFrame객체.rename(index={기존인덱스: 새인덱스})
# 열 인덱스 변경: DataFrame객체.rename(columns={기존이름: 새이름})

import pandas as pd

matrix = [
    [15, '남자', '덕영중'],
    [17, '여', '수리중']
]

df = pd.DataFrame(matrix, index=['준서','예은'], columns=['나이','성별','학교'])
print(df)
```

```python
df = df.rename(columns={
    '나이':'연령',
    '성별':'남녀',
    '학교':'소속'
})

print(df)
```

```python
df = df.rename(index={
    '준서':'학생1',
    '예은':'학생2'
})
```

```python
# 행/열 삭제

# 행 삭제: DataFrame객체.drop(행인덱스 또는 배열, axis=0, axis='index')
# 열 삭제: DataFrame객체.drop(열이름 또는 배열, axis=1, axis='columns')

import pandas as pd

exam_data = {
    '수학' : [ 90,80,70],
    '영어' : [ 98,89,95],
    '음악' : [ 95,95,100],
    '체육' : [ 100,90,90]
}

df = pd.DataFrame(exam_data, index=['서준','우현','인아'])
print(df)
df
```

```python
df2 = df.copy()
df2 = df2.drop('우현')
df2
```

```python
df3 = df.copy()
df3 = df.drop(['우현','인아'])
df3
```

```python
df3 = df.copy()
df3 = df3.drop(['영어','음악'], axis=1)
df3
```

```python
df4 = df.copy()
df4 = df4.drop(['영어','음악'], axis='columns')
df4
```

```python
# 행/열 위치 바꾸기

# 행, 열 바꾸기 : DataFrame객체.transpose()

import pandas as pd

exam_data = {
    '이름' : ['서준','우현','인아'],
    '수학' : [ 90,80,70],
    '영어' : [ 98,89,95],
    '음악' : [ 95,95,100],
    '체육' : [ 100,90,90]
}

df = pd.DataFrame(exam_data)
df
```

```python
df= df.transpose()
df
```
