```python
# reshape

import numpy as np
```

```python
arr = np.array([1,2,3,4,5,6])
print(arr.shape)
```

```python
reshaped = arr.reshape((2,3))
print(reshaped.shape)
print(reshaped)
```

```python
# astype
## 데이터타입을 변경한다

arr = np.array([1,2,3,4,5,6])
print(arr.astype(float))
```

```python
# ravel
## 다차원 배열을 1차원 배열로 변경

arr = np.array([1,2,3,4,5,6])
reshaped = arr.reshape((2,3))
print(reshaped)

raveled = reshaped.ravel()
print(raveled)
```

```python
# flatten
## 다차원 배열을 1차원 배열로 변경

arr = np.array([
    [1,2,3],
    [4,5,6]
])
print(arr)
flattened = arr.flatten()
print(flattened)
```

```python
# transpose
## 배열의 축을 바꾸는 함수

arr = np.array([
    [1,2,3],
    [4,5,6]
])

transposed = arr.transpose()
print(transposed)
```

```python
# 수학연산

arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
print(arr1 + arr2)
print(arr1 * arr2)
```

```python
# 배열간 비교연산

arr1 = np.array([5,3,3])
arr2 = np.array([1,2,6])

print(arr1 > arr2)
print(arr1 == arr2)
print(arr1 == arr1)
```

```python
# 배열 간의 논리연산

arr1 = np.array([True, False, True])
arr2 = np.array([False, True, True])

print(arr1 & arr2)
print(arr1 | arr2)
```

```python
# 배열 간의 집계연산

arr = np.array([1,2,3,4,5])
print(arr.sum())
print(arr.mean())
print(arr.min())
print(arr.max())
```

```python
arr = np.array([1,2,3,4,5])
scalar = 2
print(arr * scalar)
```
