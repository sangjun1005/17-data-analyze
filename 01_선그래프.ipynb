```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('./data/시도별_전출입_인구수.xlsx')

df.head()
```

```python
df = df.ffill()
df.head()
```

```python
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul.rename({'전입지별':'전입지'}, axis=1)
df_seoul.set_index('전입지')
df_seoul
```
