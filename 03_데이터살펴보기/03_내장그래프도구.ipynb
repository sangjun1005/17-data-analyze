```python
# 막대그래프 그리기

df = pd.read_excel('./data/남북한발전전력량 (1).xlsx')

df_ns = df.iloc[[0,5],3:]

df_ns.index= ['South','North']

df_ns = df_ns.astype(int)

tdf_ns = df_ns.T

print(tdf_ns.head())
print('\n')
tdf_ns.plot(kind='bar');
```

```python
# 가로 막대 그래프

tdf_ns.plot(kind='barh');
```

```python
# 누적 막대 그래프

tdf_ns.plot(kind='bar', stacked=True);
```

```python
# 히스토그램

tdf_ns.plot(kind='hist');
```

```python
# 차트 서식 결정

tdf_ns.plot(kind='hist',
            alpha=0.5,           # 투명도 (0-1)
            figsize=(6,2),       # 그림 크기 (가로, 세로)
            grid=True,           # 그리드 표시 여부
            color='skyblue',     # 차트 색
            edgecolor='black')   # 테두리 색
```

```python
tdf_ns.hist(figsize=(6,2));
```

```python
tdf_ns.hist(bins=20, grid=False, figsize=(6,2));
```
